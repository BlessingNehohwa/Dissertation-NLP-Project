### Data Cleaning and Data Preprocessing
1. We need to lowercase all the words
2. Removing all stop words
3. Tokenize the text
4. Lemmatization perfomance
5. Non-alphabetical characters removal from the text

```python
#import nltk
#nltk.download('punkt')

#import nltk
#nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
#creating list of possible stopwords from nltk library
stop = stopwords.words('english')
def cleanText(txt):
  # lowercasing
    txt = txt.lower()
    # tokenization
    words = nltk.word_tokenize(txt)
  # removing stopwords & mennatizing the words
    words = ' '.join([lemma.lemmatize(word) for word in words if word not in (stop)])
    text = "".join(words)
  # removing non-alphabetic characters
    txt = re.sub('[^a-z]',' ',text)
    return txt  
#applying cleantweet function on tweet text column
findata_final['cleaned_text'] = findata_final['text'].apply(cleanText)
findata_final.head(10)
```

- Cleaned the text
- Created a separate column named cleaned_text

```python
### Feature and target variable creation
X=findata_final.cleaned_text
y = findata_final.label
```

### Train Test Split
- We will divide the dataset into train and test set in the ratio of 80:20 i.e., 80% for training the machine learning model and 20% for testing the model.
- Stratification application so we can maintain the balanced proportion of sentiments in both training and test set 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20,stratify=y, random_state=0)
```

### Bag of words(BOW) Vectorisation
- We are using the count Vectorizer method to vectorise our textual data

```python
count_vectorizer = CountVectorizer()
count_train = count_vectorizer.fit_transform(X_train)
count_test = count_vectorizer.transform(X_test)
```

### Term Frequency - Inverse Document Frequency (TF-IDF) Vectorisation
- We using TF-IDF vectoriser to vectorise data to enable us to compare the performance of machine learning model based on TF-IDF and BOW.
- The two parameters we are using are max_df=0.8 and ngram_range=(1.2).
> max_df=0.8 restricts the vectorizer to use the words which have more than 80% frequency in the document while the ngra_range parameter allows both the unigrams and the bigrams in the textual data.

```python
# Bigrams
tfidf_vectorizer = TfidfVectorizer( max_df=0.8, ngram_range=(1,3))
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

## Building Machine Learning Models
- We are going to build two machine learning models on both BOW and TF-IDF vectorised data
1. Multinomial Naive  Bayes
2. Support Vector Machine 

### Fitting the machine learning models

```python
# Multinomial Naive bayes bag of words model
mnb_bow = MultinomialNB()
mnb_bow.fit(count_train, y_train)
# Multinomial Naive bayes tf-idf  model
mnb_tfidf = MultinomialNB()
mnb_tfidf.fit(tfidf_train, y_train)
#SVM bag of words model
svm_bow =SVC(probability=True,kernel='linear')
svm_bow.fit(count_train, y_train)
# SVM tf-idf model
svm_tfidf = SVC(probability=True,kernel='linear')
svm_tfidf.fit(tfidf_train, y_train)
```

### Cross validation
- Here we are splitting the training data into multiple folds and train the model on those fold by setting one fold data for evaluation purposes.
- This is critical for Overfitting pattern detection in the data

```python
# 10-folds cross validation
kfold = KFold(n_splits=15)
scoring = 'accuracy'
acc_mnb = cross_val_score(estimator = mnb_bow, X = count_train, y = y_train, cv = kfold,scoring=scoring)
acc_svm = cross_val_score(estimator = svm_bow, X = count_train, y = y_train, cv = kfold,scoring=scoring)
acc_mnb_tfidf = cross_val_score(estimator = mnb_tfidf, X = tfidf_train, y = y_train, cv = kfold,scoring=scoring)
acc_svm_tfidf = cross_val_score(estimator = svm_tfidf, X = tfidf_train, y = y_train, cv = kfold,scoring=scoring)
# compare the average 10-fold cross-validation accuracy
crossdict = {        
                'MNB-BoW': acc_mnb.mean(),
               
                'SVM-BoW':acc_svm.mean(),
               
                'MNB-tfidf': acc_mnb_tfidf.mean(), 
               
                'SVM-tfidf': acc_svm_tfidf.mean(),
             
                }
cross_df = pd.DataFrame(crossdict.items(), columns=['Model', 'Cross-val accuracy'])
cross_df = cross_df.sort_values(by=['Cross-val accuracy'], ascending=False)
cross_df
```

```python
def evaluate_model(model,model_name,test_set,model_results):
    pred = model.predict(test_set)
    acc= accuracy_score(y_test, pred)
    prec = precision_score(y_test, pred,pos_label='positive',
                                               average='macro')
    rec = recall_score(y_test, pred,pos_label='positive',
                                               average='macro')
    f1 = f1_score(y_test, pred,pos_label='positive',
                                               average='macro')
    cm = confusion_matrix(y_test, pred, labels=['positive', 'negative','neutral'])
   
    # plot_confusion_matrix(cm, classes=['positive', 'negative','neutral'])
    cmd = ConfusionMatrixDisplay(confusion_matrix =cm,display_labels=['positive', 'negative','neutral'])
    cmd.plot()
    results =pd.DataFrame([[model_name,acc, prec,rec,f1]],
                   columns = ['Model', 'Accuracy','Precision', 'Recall', 'F1 Score'])
    model_results = model_results.append(results, ignore_index = True)
    return model_results
```

```python
model_results = evaluate_model(svm_bow,'SVM-BoW',count_test,model_results)
model_results
```

### Prediction :  SVM-TF-IDF

- We are doing our prediction on real time data, therefore we need to preprocess the data and then transform it, into vectorized format using TF-IDF Vdectorizer.


```python
def predict_text(lst_text,model):
    df_test = pd.DataFrame(lst_text, columns = ['test_sent'])
    # apply data preprocessing
    df_test["test_sent_cleaned"] = df_test["test_sent"].apply(cleanText)
    # transforming text into tf-idf vectorized format
    tfidf_bigram = tfidf_vectorizer.transform(lst_text)
    # model prediction
    prediction = model.predict(tfidf_bigram)
    # saving prediction in dataframe by creation prediction column
    df_test['prediction']=prediction
    # subset dataframe to include only original entences and model prediction
    df_test = df_test[['test_sent','prediction']]
    return df_test

sentences = [
        "Both the net sales and operating profit were record high in the company 's history",
        "According to the company , in addition to normal seasonal fluctuation the market situation has weakened during autumn 2008 ",
        "Cramo Plc is a service company specialising in construction machinery and equipment rental and rental-related services  as well as rental and sale of modular space"
  ]
predict_text(sentences,svm_tfidf)
```
